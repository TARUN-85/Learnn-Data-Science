# IMAGE PROCESSING OF DOG-PHOTO

#### IMPORTING NUMPY & MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
felix  = np.array(plt.imread('dog.JPG'))
felix.shape
```

```python
from PIL import Image
im = Image.fromarray(felix)
im.save("your_file.jpeg")
```

```python
plt.imshow(felix)
```

# 1.CROPPING

```python
#SLICING IS USED TO CROP IT
plt.imshow(felix[:,500:,:])
```

# 2.FLIPPING IMAGE

```python
# FLIP ALONG VERTICAL AXIS 
plt.imshow(felix[:,:500:-1,:])
```

```python
# FLIP ALONG HORIZONTAL AXIS 
plt.imshow(felix[::-1,500::,:])
```

```python
# FLIP ALONG DEPTH: IT WILL FLIP ALONG CHANNELS R:G:B
plt.imshow(felix[:,::,::-1])
```

# 3. BLURRING IMAGE

```python
#BLURRING IN IMAGE COMES WHEN WE INCREASE STEP SIZE
plt.imshow(felix[::50,500::50,:])
```

# 4.CONTRASTING IMAGE

```python
A = np.where(felix > 150,100,0)
plt.imshow(A)
```

# 5.NEGATIVE IMAGES

```python
bw = 255-felix
plt.imshow(bw[:,500:,:])
```

# 6.CHANNEL

```python
plt.imshow(felix[:,500:,0])
```

# 7. HIDING FACE

```python
felix  = np.array(plt.imread('dog.JPG'))
felix[:] = 0
plt.imshow(felix[:,500:,:])
```
